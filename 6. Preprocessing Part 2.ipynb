```python
def build_dictionary(words):
    global dictionary
    data_list = list()
    for word in words:
        if word in dictionary:
            data_list.append(dictionary[word])
        else:
            dictionary[word] = len(dictionary)
            data_list.append(dictionary[word])
    return data_list
```
